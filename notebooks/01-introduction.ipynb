```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
amplitude = 1.5
omega = np.pi / 2
phi = 0.5 * np.pi

X = np.linspace(-2 * np.pi,  2 * np.pi, 100)
y = amplitude * np.cos(omega * X + phi)

fig, ax = plt.subplots(figsize=(10,5))
ax.grid(color='#eeeeee', linestyle='-', linewidth=1)
ax.set_xticks(
    [-2.0 * np.pi, 
     -1.5 * np.pi,
     -1.0 * np.pi, 
     -0.5 * np.pi, 
      0.0, 
      0.5 * np.pi, 
      1.0 * np.pi, 
      1.5 * np.pi, 
      2.0 * np.pi])
ax.set_xticklabels([
    r"$-2\pi$",
    r"$-\frac{3}{2}\pi$", 
    r"$-\pi$",
    r"$-\frac{1}{2}\pi$",
    "$0$", 
    r"$\frac{1}{2}\pi$",
    r"$\pi$", 
    r"$\frac{3}{2}\pi$", 
    r"$2\pi$"])
ax.plot(X, y)
```

## Discrete-time signals

### Delta Signal

The delta signal (denoted by $\delta[n]$) is a signal where sample number zero has a value of one, while all other samples have a value of zero.

```python
x = np.arange(-10, 10)
y = list(map(lambda val: 1 if val == 0 else 0, x) )
fig, ax = plt.subplots(figsize=(10,5))
ax.set_xticks([-10, -5, 0, 5, 10])
ax.stem(x, y, basefmt=' ')
```

### Unit Signal

The unit signal (denoted by $u[n]$)

```python
x = np.arange(-10, 10)
y = list(map(lambda val: 1 if val >= 0 else 0, x) )
fig, ax = plt.subplots(figsize=(10,5))
ax.set_xticks([-10, -5, 0, 5, 10])
ax.stem(x, y, basefmt=' ')
```

### Exponential Decay

```python
x = np.arange(-17, 17)
a = 0.8
y = list(map(lambda val: (a ** val) if val >= 0 else 0, x) )
fig, ax = plt.subplots(figsize=(13,5))
ax.set_xticks([-15, -10, -5, 0, 5, 10, 15])
ax.stem(x, y, basefmt=' ')
```
